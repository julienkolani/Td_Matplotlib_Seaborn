## Introduction

**Matplotlib** est l'une des bibliothèques les plus populaires pour la visualisation de données en Python. Elle a été développée en 2003 par **John Hunter**, un neurobiologiste américain qui cherchait à reproduire les capacités graphiques de MATLAB dans un environnement open-source. Cette démarche a conduit à la naissance d’un outil qui a révolutionné la façon dont les scientifiques et ingénieurs visualisent leurs données.

## Fonctionnement général

Matplotlib propose deux approches principales pour créer des graphiques :

1. **Pyplot** : Une interface de haut niveau inspirée de MATLAB, qui permet de générer des graphiques rapidement avec une syntaxe simple.
2. **API orientée objets** : Une approche plus flexible, adaptée pour des visualisations complexes, où chaque composant (Figure, Axes, etc.) est manipulé individuellement.

---

## Structure de Matplotlib

### Composants clés

- **Figure** : Le conteneur principal qui regroupe tous les éléments d’un graphique. On peut la comparer à une "toile".
- **Axes** : La zone d’un graphique où les données sont tracées (ne pas confondre avec les axes X et Y). Une Figure peut contenir plusieurs Axes.
- **Éléments graphiques** : Les lignes, les points, les barres, les annotations, les titres, etc., qui composent un graphique.
- **Artist** : L'objet de base pour tous les éléments visuels dans Matplotlib. Les textes, lignes et autres éléments graphiques héritent de cette classe.

### Hiérarchie de Matplotlib
Matplotlib suit une structure hiérarchique claire :
1. **Figure** : Conteneur principal.
2. **Axes** : Graphiques individuels dans une Figure.
3. **Éléments graphiques** : Les données visualisées, comme des lignes, des barres ou des points.

# Pyplot

Pyplot est une interface de type MATLAB fournie par Matplotlib pour créer des graphiques de manière simple en utilisant une approche procédurale :

- **Fonctionnalité** principale : chaque fonction pyplot modifie la figure ou les axes actifs.
- **Notion d'état** : Pyplot garde en mémoire l’état de la figure active et des axes, ce qui permet de tracer successivement sans redéfinir ces éléments.

````python
# Importation du module pyplot de matplotlib pour la création de graphiques
import matplotlib.pyplot as plt

# Tracer de la courbe en utilisant uniquement les valeurs de Y
# La fonction plt.plot() génère automatiquement l'axe X basé sur l'index des valeurs de Y
# ici, les valeurs de Y sont fournies sous forme de liste : [1, 9, 3, 4, 1, 9, 3, 4]
# L'axe X sera donc constitué des indices associés aux valeurs de Y (de 0 à 7 pour cette liste)
plt.plot([1, 9, 3, 4, 1, 9, 3, 4])

# Définir un titre pour l'axe des ordonnées (l'axe Y)
# Ici, on indique que l'axe Y représente les valeurs de Y
plt.ylabel('Valeur de Y')

# Afficher le graphique à l'écran
# La fonction plt.show() permet de visualiser le graphique généré
plt.show()
````

Par défaut :

    Les valeurs X vont de 0 à n-1 où n est la longueur de Y.
    Ici, X = [1, 9, 3, 4, 1, 9, 3, 4].

### Tracer plusieurs courbes dans un seul appel :

````python
# Importation de la bibliothèque numpy pour manipuler les tableaux numériques
import numpy as np

# Création d'un tableau de valeurs t allant de 0 à 5, avec un pas de 0.2
# np.arange() génère une séquence d'éléments allant de 0 à 5 (non inclus), espacés de 0.2
t = np.arange(0., 5., 0.2)

# Tracer plusieurs courbes sur le même graphique avec différentes couleurs et styles de ligne
# Chaque appel à plt.plot() trace une courbe distincte avec un style spécifié
plt.plot(t, t, 'r--',   # 'r--' : ligne rouge avec des tirets
         t, t**2, 'bs',  # 'bs' : carrés bleus pour la courbe t^2
         t, t**3, 'g^')  # 'g^' : triangles verts pour la courbe t^3

plt.show()
````

````python
# Définition des noms des catégories et des valeurs associées
names = ['A', 'B', 'C']
values = [1, 10, 100]

# Définir la taille de la figure (largeur, hauteur)
# Ici, la taille de la figure est de 15 pouces de large et 3 pouces de haut
plt.figure(figsize=(15, 3))

# Premier sous-graphique : Diagramme à bâton
# plt.subplot() crée une grille de sous-graphes et définit où chaque graphique sera placé
# (1 ligne, 3 colonnes, et on sélectionne le premier sous-graphique)
plt.subplot(131)
plt.bar(names, values)  # Trace un diagramme à bâton (barres verticales) avec les données
plt.title('Diagramme à bâton')  # Ajoute un titre au premier sous-graphique

# Deuxième sous-graphique : Diagramme de dispersion
# (1 ligne, 3 colonnes, et on sélectionne le deuxième sous-graphique)
plt.subplot(132)
plt.scatter(names, values)  # Trace un nuage de points avec les données
plt.title('Diagramme de dispersion')  # Ajoute un titre au deuxième sous-graphique

# Troisième sous-graphique : Graphique linéaire
# (1 ligne, 3 colonnes, et on sélectionne le troisième sous-graphique)
plt.subplot(133)
plt.plot(names, values)  # Trace un graphique linéaire reliant les points (nom, valeur)
plt.title('Graphique linéaire')  # Ajoute un titre au troisième sous-graphique

# Si désiré, on peut ajouter un titre global pour tous les sous-graphes (commenté ici)
# plt.suptitle('Graphiques catégoriels')  # Ajoute un titre général à la figure

# Afficher tous les graphiques générés
plt.show()
````

````python
# Création d'un tableau t allant de 0 à 5, avec un pas de 0.01
# Cela permet de générer des points de données très proches pour une courbe lisse
t = np.arange(0.0, 5.0, 0.01)

# Calcul de la fonction cosinus de 2πt, créant une courbe cosinus
# La fonction cos() prend un argument en radians, et 2πt est donc utilisé pour obtenir une oscillation complète
s = np.cos(2 * np.pi * t)

# Tracer la courbe cosinus
plt.plot(t, s)  # La courbe représente cos(2πt), une fonction périodique

# Ajouter une annotation avec une flèche pointant vers un maximum local de la courbe
# 'xy' est la position du point annoté (ici, on choisit un point autour de (2, 1))
# 'xytext' est la position du texte annoté (ici, légèrement décalée vers la droite et le haut)
# 'arrowprops' est un dictionnaire qui permet de personnaliser l'apparence de la flèche
plt.annotate('Maximum local',
             xy=(2, 1),  # Position du maximum local sur la courbe
             xytext=(3, 1.5),  # Position du texte annoté
             arrowprops=dict(facecolor='black', shrink=0.05))  # Flèche noire avec réduction de taille

# Définir les limites de l'axe des ordonnées (y) pour zoomer sur la courbe
# Ici, les valeurs de y sont limitées entre -2 et 2 pour mieux visualiser les oscillations
plt.ylim(-2, 2)

# Afficher le graphique avec la courbe cosinus et l'annotation
plt.show()
````

### Tracer et manipuler des images avec Matplotlib

Chargement et affichage d'une image

Nous commencerons par télécharger une image depuis une URL et la convertir en tableau NumPy à l'aide de Pillow.

````python
# Importation des modules nécessaires
import requests  # Pour télécharger l'image via URL
from PIL import Image  # Pour ouvrir et manipuler l'image avec Pillow
from io import BytesIO  # Pour traiter l'image téléchargée en mémoire
````

````python
# URL de l'image
image_url = "https://cdn.pixabay.com/photo/2024/06/24/04/05/woman-8849047_1280.jpg"

# Télécharger l'image à partir de l'URL
response = requests.get(image_url)  # Effectue une requête GET pour récupérer l'image
response.raise_for_status()  # Vérifie si la requête a réussi (lance une exception si une erreur survient)

# Charger l'image dans un objet Pillow (Image) à partir des données binaires téléchargées
img = Image.open(BytesIO(response.content))  # Ouvre l'image en mémoire en utilisant BytesIO

# Convertir l'image en un tableau NumPy pour pouvoir la manipuler facilement
img_array = np.asarray(img)  # Convertit l'objet Image en un tableau NumPy

# Dimensions de l'image (hauteur, largeur et nombre de canaux)
print("\nDimensions de l'image :", img_array.shape)  # Affiche la forme du tableau (dimensions)
````

Affichage de l'image originale

````python
# Tracer l'image
plt.imshow(img_array)
plt.title("Image")
#plt.axis("off")
plt.show()
````

````python
from scipy.ndimage import gaussian_filter  # Pour appliquer un filtre gaussien

# Application d'un filtre gaussien à une matrice NumPy
# On peut appliquer un filtre gaussien à une matrice NumPy, ce qui est très utile en traitement d'images ou en analyse de données.
# Le filtre gaussien permet de lisser les données et de réduire le bruit.
# Par exemple, dans le cas d'une image, il permet de flouter l'image en réduisant les détails fins.

# Exemple  : créer une matrice 5x5
a = np.arange(50, step=2).reshape((5,5))  # Crée une matrice 5x5 avec des valeurs de 0 à 48
print("Matrice originale :\n", a)

# Appliquer un filtre gaussien à la matrice
a_filtered = gaussian_filter(a, sigma=1)  # sigma détermine l'intensité du flou
print("\nMatrice après filtre gaussien :\n", a_filtered)
````

````python
# L'application d'un filtre gaussien lisse la matrice, ce qui est visible par le changement des valeurs.

# Appliquer un filtre gaussien à l'image pour flouter l'image et réduire les détails
img_filtered = gaussian_filter(img_array, sigma=2)  # sigma détermine l'intensité du flou

# Afficher l'image originale et l'image filtrée côte à côte
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # Créer un sous-graphique avec 2 colonnes
ax[0].imshow(img_array)  # Affiche l'image originale
ax[0].set_title("Image originale")
ax[0].axis('off')  # Désactive les axes pour une présentation plus propre

ax[1].imshow(img_filtered)  # Affiche l'image après le filtre gaussien
ax[1].set_title("Image filtrée (gaussienne)")
ax[1].axis('off')  # Désactive les axes

# Affichage des images
plt.show()
````

# Seaborn

## Introduction

**Seaborn** est une bibliothèque Python basée sur **Matplotlib** et conçue pour la visualisation statistique des données. Développée par **Michael Waskom** en 2012, Seaborn offre des outils simples pour créer des graphiques complexes tout en améliorant l'esthétique par défaut, rendant les visualisations plus compréhensibles et attrayantes. Contrairement à Matplotlib, qui nécessite une gestion détaillée des axes et des éléments graphiques, Seaborn permet de créer des visualisations directement à partir de structures de données comme les **DataFrames** de **pandas**.

### Historique
- **2012** : Création de Seaborn par Michael Waskom, dans le but d'améliorer la visualisation des données statistiques en Python, en utilisant la puissance de Matplotlib mais avec une approche plus accessible.
- **2013** : La version initiale de Seaborn devient rapidement populaire auprès des utilisateurs de pandas grâce à sa simplicité d’utilisation et à son style graphique élégant.
- **Aujourd'hui** : Seaborn est une bibliothèque largement utilisée dans la communauté de la data science et de l’analyse statistique pour sa facilité d’utilisation et ses capacités avancées de visualisation.

## Objectifs de Seaborn

Seaborn permet de :

1. **Simplifier la création de graphiques statistiques** : Permet de créer des graphiques complexes avec moins de code, idéal pour des analyses exploratoires rapides.
2. **Faciliter l'intégration avec pandas** : Fonctionne de manière transparente avec les **DataFrames** de pandas, ce qui simplifie le processus de visualisation de données structurées.
3. **Produire des graphiques esthétiques** : Seaborn offre une large gamme de couleurs et de styles de graphiques par défaut, garantissant des visualisations attractives et informatives.

### Les caractéristiques clés de Seaborn

- **Visualisations statistiques** : Seaborn est principalement axé sur les graphiques qui permettent de comprendre la structure statistique des données.
- **Simplicité et clarté** : Contrairement à Matplotlib, qui nécessite une configuration détaillée des éléments du graphique, Seaborn permet de créer des graphiques complexes avec peu de code, tout en restant flexible.

### Qu'est-ce qu'une graine (seed) ?
Une graine est une valeur de départ utilisée par les algorithmes de génération de nombres pseudo-aléatoires (PRNG - Pseudo-Random Number Generators). Elle sert de point de référence pour initialiser le générateur.

Les nombres générés ne sont pas vraiment aléatoires, mais calculés à partir d'algorithmes déterministes.
Si on initialise le générateur avec la même graine, les nombres produits seront identiques à chaque exécution.
Pourquoi "pseudo" ?
Le terme pseudo-aléatoire signifie que les nombres produits semblent aléatoires, mais ils suivent une séquence déterminée par la graine et l'algorithme. Le caractère aléatoire est donc simulé.

Un PRNG utilise :

Une graine initiale (seed) : Point de départ de la séquence.
Un algorithme mathématique : Transforme la graine en une suite de valeurs.
Les étapes typiques :

La graine est entrée dans un algorithme, souvent basé sur des opérations arithmétiques.
L'algorithme produit une séquence de nombres (par exemple, entre 0 et 1).
Chaque appel au générateur donne un nouvel élément de la séquence.
Exemple simplifié : Générateur linéaire congruentiel.

En fixant une graine, on peut reproduire exactement les mêmes résultats à chaque exécution.
Important dans les expériences scientifiques ou les tests logiciels.
Exemple : Tester un modèle d'apprentissage automatique sur un ensemble fixe de données aléatoires.

Une graine fixe permet de reproduire les erreurs aléatoires, facilitant leur détection et correction.
Comparaison des performances :

Si différents algorithmes doivent être comparés, utiliser la même graine garantit qu'ils sont testés dans des conditions identiques.

````python
# Bibliothèque pour la manipulation de données sous forme de DataFrame
import pandas as pd
import seaborn as sns

# Génération de données aléatoires
np.random.seed(42)  # Fixe la graine du générateur aléatoire pour assurer la reproductibilité des résultats

# Définition des jours de la semaine et du nombre de lignes
jours = ['Thur', 'Fri', 'Sat', 'Sun']  # Liste des jours possibles
n = 200  # Nombre total de lignes (observations) à générer dans le dataset

# Paramètres de la distribution normale pour la colonne 'total_bill'
mu_total_bill = 30  # Moyenne de la distribution des montants totaux des factures ('total_bill')
sigma_total_bill = 10  # Ecart-type de la distribution des montants totaux des factures

# Paramètres de la distribution normale pour la colonne 'tip'
mu_tip = 5  # Moyenne de la distribution des pourboires ('tip')
sigma_tip = 2  # Ecart-type de la distribution des pourboires

# Génération des données
data = {
    # Génère des montants totaux de factures selon une distribution normale
    # La fonction np.random.normal crée un tableau de n valeurs suivant une loi normale définie par mu et sigma
    # np.round arrondit chaque valeur à 2 décimales pour simuler un montant réaliste
    'total_bill': np.round(np.random.normal(mu_total_bill, sigma_total_bill, n), 2),

    # Génère des pourboires selon une distribution normale avec des valeurs arrondies à 2 décimales
    'tip': np.round(np.random.normal(mu_tip, sigma_tip, n), 2),

    # Attribue un jour aléatoire pour chaque observation en choisissant dans la liste des jours
    # np.random.choice permet de tirer n échantillons parmi les jours possibles
    'day': np.random.choice(jours, n),

    # Génère un entier aléatoire entre 1 et 5 (inclus) pour simuler le nombre de personnes dans un groupe
    'size': np.random.randint(1, 6, n)
}

# Création du DataFrame Pandas à partir du dictionnaire 'data'
df = pd.DataFrame(data)

# Affiche les 5 premières lignes du DataFrame pour inspection
print(df.head())
````

````python
# Tracer un histogramme avec la courbe de densité
sns.histplot(df['total_bill'], kde=True)  # kde=True ajoute une estimation de densité par noyau (courbe lisse)
plt.title('Distribution des Montants de Facture')  # Ajouter un titre
plt.xlabel('Montant de la Facture ($)')  # Étiqueter l'axe X
plt.ylabel('Fréquence')  # Étiqueter l'axe Y
plt.show()
````

````python
# Les Boîtes à moustaches
sns.boxplot(x='day', y='tip', data=df)  # Création d'un boxplot pour visualiser la distribution des pourboires (y) selon les jours (x).


# Une boîte à moustaches résume la distribution des données :
# - La ligne centrale dans la boîte représente la médiane (valeur centrale des données triées).
# - Les bords de la boîte indiquent le premier quartile (Q1, 25%) et le troisième quartile (Q3, 75%).
# - Les "moustaches" s'étendent jusqu'aux valeurs minimales et maximales dans une plage définie comme [Q1 - 1.5*IQR, Q3 + 1.5*IQR], où IQR est l'écart interquartile (Q3 - Q1).
# - Les points en dehors des moustaches sont considérés comme des valeurs aberrantes (outliers).


plt.title('Boxplot des Pourboires par Jour')
plt.xlabel('Jour')
plt.ylabel('Pourboire ($)')
plt.show()
````

Les bibliothèques **Seaborn** et **Matplotlib** acceptent des données dans plusieurs formats, mais elles ont des préférences spécifiques en fonction du type de visualisation que tu veux réaliser. Voici un récapitulatif des types de données acceptés :

---

Seaborn est conçu pour fonctionner principalement avec des données tabulaires et utilise des structures bien adaptées aux pandas.

#### **Formats principaux :**
- **DataFrame (pandas)** :
  - Le format privilégié par Seaborn.
  - Les colonnes du DataFrame peuvent être référencées directement dans les arguments `x`, `y`, ou autres (par exemple, `hue`, `size`, `style`).

- **Series (pandas)** :
  - Les colonnes d'un DataFrame peuvent être passées individuellement si elles sont des pandas Series.

#### **Formats alternatifs :**
- **Numpy Arrays** :
  - Les tableaux NumPy sont acceptés, mais il faut fournir explicitement les valeurs pour `x` et `y`.
  - Exemple :
    ```python
    import seaborn as sns
    import numpy as np
    x = np.random.randn(100)
    y = np.random.randn(100)
    sns.scatterplot(x=x, y=y)
    ```

- **Dictionnaires** :
  - Si les données sont passées sous forme de dictionnaire avec des listes comme valeurs, elles seront converties en DataFrame par Seaborn.

#### **Formats non pris en charge directement** :
- Les formats non tabulaires comme les listes imbriquées, les tuples, ou les formats orientés séries temporelles nécessitent une conversion préalable.

---

### **2. Types acceptés par Matplotlib :**

Matplotlib est plus bas niveau que Seaborn et donc plus flexible, mais moins automatisé.

#### **Formats principaux :**
- **Listes Python** :
  - Les listes simples ou imbriquées peuvent être utilisées.
  - Exemple :
    ```python
    import matplotlib.pyplot as plt
    x = [1, 2, 3, 4]
    y = [10, 20, 25, 30]
    plt.plot(x, y)
    plt.show()
    ```

- **Numpy Arrays** :
  - Les tableaux NumPy sont souvent utilisés car Matplotlib est optimisé pour ce format.
  - Exemple :
    ```python
    import numpy as np
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.show()
    ```

- **Pandas Series ou DataFrame** :
  - Bien qu'il n'y ait pas d'intégration directe, Matplotlib fonctionne bien avec les pandas Series ou DataFrame. Les index pandas peuvent également être utilisés pour les axes.

#### **Formats alternatifs :**
- **Dictionnaires** :
  - Les clés peuvent être utilisées comme labels et les valeurs comme données.
  - Exemple :
    ```python
    data = {'x': [1, 2, 3], 'y': [10, 20, 30]}
    plt.scatter(data['x'], data['y'])
    plt.show()
    ```

- **Python Generators** :
  - Les générateurs Python sont pris en charge, bien que leur utilisation soit plus rare.
  
---

### **Comparaison Seaborn vs. Matplotlib :**

| Fonctionnalité                  | Seaborn                      | Matplotlib                  |
|----------------------------------|------------------------------|-----------------------------|
| Préférence de format            | DataFrame / Series (pandas)  | Listes / Numpy Arrays       |
| Simplification de code          | Très automatisé              | Plus manuel                 |
| Compatibilité avec des données brutes | Limité à tabulaire         | Très flexible               |

### Visualisation des données pendant l'Analyse en Composantes Principales (ACP) avec le PIB et les Indicateurs de Gouvernance Mondiale (WGI)

Nous allons maintenant passer à la dernière phase du TD : l'utilisation de la visualisation dans un cas pratique d'Analyse en Composantes Principales (ACP). Dans ce cas, nous allons travailler avec un jeu de données portant sur les **Indicateurs de Gouvernance Mondiale (WGI)**, qui inclut des informations sur la qualité de la gouvernance dans plus de 200 pays, sur la période 1996-2023.

Les **WGI** sont basés sur des données collectées par plus de 30 organisations internationales, des think tanks, des ONG et des entreprises privées. Ces données couvrent six dimensions de la gouvernance, et sont compilées à partir de perceptions exprimées par des milliers de répondants et d'experts à travers le monde. Les six dimensions sont les suivantes :

1. **Voice and Accountability** (Expression libre et responsabilité)
2. **Political Stability and Absence of Violence** (Stabilité politique et absence de violence)
3. **Government Effectiveness** (Efficacité du gouvernement)
4. **Regulatory Quality** (Qualité de la régulation)
5. **Rule of Law** (État de droit)
6. **Control of Corruption** (Contrôle de la corruption)

### Objectif de l'ACP dans ce contexte :
L'ACP va être utilisée ici pour réduire la dimensionnalité des données WGI et nous aider à identifier des relations entre les différentes dimensions de la gouvernance et les pays. En visualisant les données projetées sur un espace réduit (2D ou 3D), nous serons en mesure de mieux comprendre les groupes de pays ayant des caractéristiques de gouvernance similaires et d'identifier les tendances globales à travers les différentes dimensions.

### Étapes de l'analyse :
1. **Prétraitement des données :**
   - **Chargement et exploration des données :** Nous commencerons par charger le jeu de données WGI et examiner les différentes dimensions disponibles pour chaque pays.
   - **Nettoyage des données :** Nous vérifierons la présence de valeurs manquantes et procéderons, si nécessaire, à une imputation ou une suppression des données manquantes.
   - **Normalisation des données :** Avant de procéder à l'ACP, nous normaliserons les données pour que chaque dimension ait la même échelle. Cela est crucial pour que les résultats de l'ACP ne soient pas biaisés par des différences d'échelle entre les variables.

2. **Application de l'ACP :**
   - Nous utiliserons l'ACP pour réduire le nombre de dimensions de ces données, généralement en conservant les deux premières composantes principales, qui expliqueront la majeure partie de la variance des données.

3. **Visualisation des résultats :**
   - Après l'ACP, nous afficherons les résultats sous forme de nuage de points, où chaque pays sera représenté par un point dans l'espace réduit, avec sa position déterminée par les deux premières composantes principales.
   - Nous pourrons colorier les points en fonction de certaines caractéristiques des pays (par exemple, par région géographique ou niveau de développement économique), afin de mieux comprendre les relations entre la gouvernance et les autres facteurs.

````python
# URL brute du fichier Excel
url = "https://raw.githubusercontent.com/julienkolani/Td_Matplotlib_Seaborn/main/wgidataset.xlsx"

# Charger le fichier Excel directement depuis l'URL
df = pd.read_excel(url)

# Afficher la taille du fichier
print("Taille du fichier : ", df.shape)

# Afficher les 10 premières lignes
df.head(10)
````

````python
print("Les colonnes du Dataset")
print(df.columns.values)

# Suppression des lignes avec des valeurs manquantes
df = df.replace("..", pd.NA)  # Remplace les valeurs de type ".." (qui sont souvent utilisées pour désigner des valeurs manquantes) par des NaN (Not a Number) de pandas
df.dropna(inplace=True)  # Supprime les lignes contenant des valeurs manquantes (NaN). inplace=True modifie le DataFrame d'origine sans retourner une nouvelle copie

# Affiche les 5 premières lignes du DataFrame après la suppression des valeurs manquantes
df.head(5)
````

````python
print("La nouvelle valeur du dataset est : ", df.shape)
y_values = df.iloc[:, 0:5]
print("Variables considérées comme catégorielles")
y_values.head(5)
````

````python
print("Variables considerer comme valeurs X")

x_values = df.iloc[:, 5:]
# Afficher les 5 premières lignes des données de x_values pour un aperçu
x_values.head(5)

# Convertir toutes les valeurs de x_values en type float
x_values = x_values.astype(float)  # Cette ligne assure que toutes les colonnes de x_values sont de type float

print(x_values.info())

x_values  # Cette ligne renvoie les valeurs actuelles de x_values pour les examiner plus en détail
````

````python
# Importer la classe StandardScaler de la bibliothèque sklearn pour la normalisation des données
from sklearn.preprocessing import StandardScaler

# Faire une copie de x_values avant la normalisation, pour garder une version originale des données
x_values_origin = x_values.copy()  # Cette copie sera utile pour garder une référence aux colonnes originales

# Appliquer la standardisation (normalisation) des données
x_values = StandardScaler().fit_transform(x_values)  # La méthode fit_transform() ajuste le scaler aux données et les transforme en données standardisées

# Le résultat de fit_transform() est une matrice NumPy, donc il est parfois utile de vérifier son type
# type(x_values)  # Vous pouvez décommenter cette ligne pour vérifier le type de la variable x_values

# Convertir les données normalisées en DataFrame pour mieux manipuler les données
x_values = pd.DataFrame(x_values)  # Transformation du tableau NumPy en DataFrame

# Donner à la nouvelle DataFrame les noms de colonnes d'origine de x_values
x_values_scaled_df = pd.DataFrame(x_values, columns=x_values_origin.columns)  # Recréer un DataFrame avec les colonnes d'origine

# Afficher les noms des colonnes du DataFrame des données standardisées
print(x_values_scaled_df.columns)  # Cela permet de vérifier que les colonnes sont toujours nommées comme dans les données originales
````

````python
# Trop de données ne sont pas plotables : nous allons échantillonner un sous-ensemble des données pour faciliter la visualisation

# random_state : noyaux de génération des données aléatoires pour garantir que l'échantillonnage est reproductible
x_values_random_plot = x_values.sample(frac=0.012, random_state=42)  # Sélectionne 1.2% des données de manière aléatoire

# Pour avoir des plots un peu lisibles, car la grande quantité de données empêche la bonne visualisation
print(x_values_random_plot.shape)  # Affiche la forme (dimensions) de l'échantillon pour vérifier sa taille

# Afficher un pairplot des données échantillonnées
sns.pairplot(x_values_random_plot)  # Crée un pairplot, une matrice de graphiques en paires pour chaque variable
plt.show()  # Affiche le plot
````

````python
# On observe que les données présentent des corrélations, ce qui est très bon pour une ACP (Analyse en Composantes Principales)
# La corrélation indique la force et la direction de la relation linéaire entre les variables.

# Calcul de la matrice de corrélation entre les colonnes de x_values
corr_matrix = x_values.corr()  # La méthode .corr() calcule la corrélation pairwise entre toutes les colonnes numériques du DataFrame

# Affichage de la matrice de corrélation sous forme de heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)  # heatmap affiche la matrice de corrélation
# - annot=True : Ajoute les valeurs de corrélation dans chaque cellule de la heatmap
# - cmap='coolwarm' : Définit la palette de couleurs allant du bleu (corrélation négative) au rouge (corrélation positive)
# - center=0 : Centrer la couleur autour de zéro, ce qui permet de mieux distinguer les corrélations positives et négatives

# Ajouter un titre à la heatmap
plt.title('Matrice de corrélation')  # Titre de la heatmap

# Afficher la heatmap
plt.show()  # Affiche la figure générée
````

````python
# Début de l'ACP (Analyse en Composantes Principales)
from sklearn.decomposition import PCA  # Importation de la classe PCA de scikit-learn pour l'ACP

# Création de l'objet PCA
pca = PCA()  # Instanciation de l'ACP sans préciser le nombre de composants à conserver (tout sera pris en compte)

# Ajustement du modèle PCA aux données (calcul des composantes principales)
pca.fit(x_values)  # Apprentissage de l'ACP sur les données x_values

# Transformation des données originales pour les projeter sur les composantes principales
x_values_pca = pca.transform(x_values)  # Transforme les données en fonction des composantes principales obtenues par pca.fit()

# Récupération des valeurs propres associées aux composantes principales
x_values_eigenvalue = pca.explained_variance_ratio_  # Récupère le ratio de variance expliqué par chaque composante principale

# Affichage des valeurs propres (variance expliquée)
print("Valeur Propre : ", pca.explained_variance_)  # Affiche les valeurs propres, qui indiquent la variance expliquée par chaque composante

# Affichage du ratio des valeurs propres (variance expliquée relative par chaque composante)
print("\nRatio des valeurs propres : ", x_values_eigenvalue)  # Affiche le ratio de la variance expliquée par chaque composante principale
````

````python
# Création d'un DataFrame contenant les informations sur les valeurs propres et la variance expliquée par chaque composante principale
eig = pd.DataFrame(
    {
        # "Dimension" : Crée une liste de noms de dimensions (Dim1, Dim2, ...)
        "Dimension" : ["Dim" + str(x + 1) for x in range(len(x_values_eigenvalue))],

        # "Variance expliquée" : Valeurs propres, qui indiquent la quantité de variance expliquée par chaque composante principale
        "Variance expliquée" : pca.explained_variance_,

        # "% variance expliquée" : Le pourcentage de variance expliqué par chaque composante
        "% variance expliquée" : np.round(pca.explained_variance_ratio_ * 100),

        # "% cum. var. expliquée" : Le pourcentage cumulatif de la variance expliquée jusqu'à chaque composante
        "% cum. var. expliquée" : np.round(np.cumsum(pca.explained_variance_ratio_) * 100)
    }
)

# Affichage du DataFrame
eig
````

````python
# Tracer un graphique en barres de la colonne "% variance expliquée" en fonction de la "Dimension"
eig.plot.bar(x="Dimension", y="% variance expliquée", color="g")  # "g" spécifie la couleur verte pour les barres

# Afficher le graphique
plt.show()
````

````python
# Convertir les données PCA en DataFrame
x_values_pca = pd.DataFrame(x_values_pca)  # Transformation du tableau numpy x_values_pca en un DataFrame pandas

# Sélectionner un sous-ensemble aléatoire des données PCA pour une meilleure visualisation
x_values_pca_random_plot = x_values_pca.sample(frac=0.012, random_state=42)  # Échantillonnage aléatoire de 1.2% des données

# Afficher la forme (dimensions) du sous-ensemble échantillonné
print(x_values_pca_random_plot.shape)  # Affiche la taille du DataFrame après échantillonnage

# Tracer les graphiques pairplot pour examiner les relations entre les composantes principales
sns.pairplot(x_values_pca_random_plot)  # Trace les relations entre chaque paire de colonnes du DataFrame

# Afficher le graphique
plt.show()  # Affiche le pairplot
````

````python
# Calculer la matrice de corrélation des données après l'ACP
corr_matrix_pca = x_values_pca.corr()  # Calcule la corrélation entre les différentes dimensions (composantes principales)

# Afficher la matrice de corrélation sous forme de heatmap
sns.heatmap(corr_matrix_pca, annot=True, cmap='coolwarm', center=0)  # Heatmap de la matrice de corrélation avec annotations

# Afficher le graphique
plt.show()  # Affiche le heatmap de la matrice de corrélation
````

Dans ce TD, nous avons exploré plusieurs outils puissants de visualisation de données tels que **Matplotlib**, **Seaborn**, et **Plotly**, chacun ayant ses avantages, ses limitations et ses usages spécifiques.

- **Matplotlib** est l'un des outils de base pour la création de graphiques en Python. Il offre une grande flexibilité et permet de personnaliser chaque aspect d'un graphique. Cependant, sa syntaxe peut parfois être verbeuse et complexe, surtout lorsqu'il s'agit de graphiques plus avancés ou interactifs. Son principal atout réside dans sa capacité à produire des visualisations simples mais efficaces, adaptées aux besoins les plus courants.

- **Seaborn**, construit sur **Matplotlib**, ajoute un niveau de sophistication en simplifiant la création de graphiques complexes comme les cartes de chaleur, les boxplots ou les graphiques de distributions. Il facilite l'exploration des relations entre plusieurs variables, en particulier lorsqu'il s'agit de visualiser des données catégorielles et continues. Cependant, il reste limité dans la personnalisation avancée et ne permet pas autant de contrôle que **Matplotlib** dans certains cas.

- **Plotly**, en revanche, se distingue par sa capacité à créer des visualisations interactives et animées. Grâce à ses outils d'animation, il permet de visualiser l'évolution des données dans le temps, de rendre l'expérience utilisateur plus immersive et dynamique. En plus, il permet de créer des visualisations facilement partageables sur le web. Cependant, **Plotly** peut nécessiter un certain apprentissage pour bien maîtriser ses fonctionnalités avancées, et bien que son utilisation de base soit assez simple, la personnalisation fine des graphiques peut être plus complexe que dans **Matplotlib**.

### Limites de Matplotlib et Seaborn en 3D, et comment Plotly les comble

**Matplotlib** et **Seaborn** sont des outils puissants pour la visualisation de données en 2D, mais leurs capacités en 3D sont limitées. Ces bibliothèques sont principalement conçues pour des visualisations sur des plans bidimensionnels, ce qui rend la gestion de données en trois dimensions moins intuitive et plus complexe.

- **Matplotlib** propose un support pour les graphiques en 3D, mais il est relativement basique et difficile à utiliser de manière fluide pour des visualisations complexes. Par exemple, la gestion de plusieurs axes en 3D, l'interaction avec les graphiques (comme la rotation ou le zoom), et la personnalisation avancée des graphes sont limitées dans **Matplotlib**. De plus, la syntaxe nécessaire pour créer des graphiques 3D avec **Matplotlib** est souvent verbeuse et nécessite plusieurs étapes manuelles pour ajuster l'affichage.

- **Seaborn**, qui est construit sur **Matplotlib**, ne prend même pas en charge les graphiques en 3D. Il est conçu pour simplifier la création de graphiques statistiques en 2D, mais ne permet pas de créer des visualisations en 3D de manière native.

**Plotly**, quant à lui, comble ces limitations en offrant une prise en charge complète des visualisations en 3D avec une syntaxe plus intuitive et beaucoup plus de flexibilité. Grâce à **Plotly**, la création de graphiques 3D devient facile et interactive. Par exemple :

- **Support natif des graphiques 3D** : **Plotly** permet de créer des graphiques 3D comme des scatter plots, des surfaces, des contours et même des graphiques en 3D avec des animations. Cela permet d'explorer des relations complexes entre plusieurs variables en 3D de manière intuitive.

- **Interaction améliorée** : Contrairement à **Matplotlib** et **Seaborn**, **Plotly** offre une interactivité native. Les graphiques en 3D peuvent être manipulés directement dans le navigateur, avec la possibilité de tourner, zoomer et se déplacer dans la visualisation pour une exploration plus poussée des données. Cela rend l'analyse des données multidimensionnelles beaucoup plus fluide et accessible.

- **Facilité d'utilisation** : La syntaxe pour créer des graphiques 3D avec **Plotly** est bien plus simple que celle de **Matplotlib**. De plus, les paramètres sont bien plus explicites et permettent une personnalisation plus facile des axes, des couleurs, et de la disposition des points en 3D.

Bien que **Matplotlib** et **Seaborn** soient parfaits pour les visualisations en 2D, leurs fonctionnalités en 3D sont limitées. **Plotly** comble cette lacune en permettant de créer des visualisations 3D interactives avec une plus grande simplicité et flexibilité, ce qui en fait un excellent choix pour l'exploration et l'analyse de données multidimensionnelles.

### Limites et perspectives avec Plotly

L'une des principales limitations de **Plotly** est sa courbe d'apprentissage. Bien que son utilisation pour les visualisations de base soit relativement simple, la personnalisation avancée et l'intégration dans des systèmes complexes peuvent demander plus d'efforts. De plus, bien que l'interactivité soit un avantage, elle peut devenir un inconvénient dans le cas de visualisations très complexes ou de grandes quantités de données, où la réactivité peut être réduite.

### Perspectives

L'avenir des visualisations interactives avec **Plotly** semble prometteur, notamment avec la possibilité d'intégrer ces graphiques dans des dashboards interactifs, par exemple à l'aide de **Dash** (une autre bibliothèque de **Plotly**). De plus, l'intégration de **Plotly** dans des plateformes de visualisation de données en ligne, comme **Tableau** ou **Power BI**, ouvre de nouvelles perspectives pour créer des visualisations puissantes accessibles à un large public.

En somme, **Matplotlib** et **Seaborn** restent des choix robustes pour des visualisations statiques et analytiques de données. **Plotly**, quant à lui, est un excellent choix lorsque l'interactivité et l'animation sont essentielles pour explorer des données sur le long terme et dans des formats dynamiques.
